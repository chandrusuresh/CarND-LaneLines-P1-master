Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, vertices, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    xp = []
    yp = []
    xn = []
    yn = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
#             if slope > 0:#0.55 and slope < 0.75:
            if x1 > vertices[0,0,2,0] and x2 < vertices[0,0,3,0]:
                xp += [x1,x2]
                yp += [y1,y2]
#             elif slope < 0:#-0.7 and slope > -0.8:
            elif x1 > vertices[0,0,0,0] and x2 < vertices[0,0,1,0]:
                xn += [x1,x2]
                yn += [y1,y2]
            else:
                continue
#             cv2.line(img, (x1, y1), (x2, y2), color, thickness)

    if bool(xp):
        poly_coeff_pos = np.polyfit(xp,yp,1)
        #xp.sort()
        xp = [min(xp),max(xp)]
        y_pos = [int(poly_coeff_pos[0]*i + poly_coeff_pos[1]) for i in xp]
        for i in range(len(xp)-1):
            cv2.line(img,(xp[i], y_pos[i]),(xp[i+1], y_pos[i+1]), color, thickness)           
    #         cv2.line(img,(xp[i], yp[i]),(xp[i+1], yp[i+1]), color, thickness)                 
    
    if bool(xn):
        poly_coeff_neg = np.polyfit(xn,yn,1)
#         xn.sort()
        xn = [min(xn),max(xn)]
        y_neg = [int(poly_coeff_neg[0]*i + poly_coeff_neg[1]) for i in xn]        
        for i in range(len(xn)-1):
            cv2.line(img,(xn[i],y_neg[i]),(xn[i+1],y_neg[i+1]),color,thickness)
#             cv2.line(img,(xn[i],yn[i]),(xn[i+1],yn[i+1]),color,thickness)        
    
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, vertices):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines, vertices)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

# Comments
Certainly, the lane finding algorithm that I have written is fairly robust. However, in certain frames in the SolidYellowLeft example the right lane identification was way off. 

Initially, I applied a  threshold on the slope for each segmented line to only consider the slopes within my range of interest. This range of interest of the slopes lies within the slopes of the trapezoidal region of interest we had applied earlier in the algorithm.

After finding all data points within a certain range of positive slope, I applied a linear fit to these to find the best line connecting these data points. The best line connecting these data points gives us right side of the lane. Repeating the process for negative slope gives us the left side of the lane, thus completing the lane finding algorithm.

The algorithm was very sensitive to the range of slopes applied. The narrower the range, the closer the predicted lane is to the actual lane. But fining this narrower range is time consuming and may not be robust as the road condition/curvature changes.
This can be corrected by applying a moving average filter over the previous n (say 5) frames.

I then scraped the criteria for the slope. Instead, for the left lane, I looked at the datapoints that correspond to the range of values in my left transverse edge of the trapezoid. and determined the best fit. The same procedure was followed for the right lane marker. This procedure was found to be more robust than the first one described above.

The algorithm might fail when the road surface/curvature changes. Note that we are strictly looking for lines. So when the road has sharp curves, the algorithm is likely to perform poorly. Again, this can be somewhat overcome by applying a higher order polynomial fit.

By applying strict ranges on the slope, the algorithm is heavily dependent on the sensor fixed in its position rigidly. Even a slight change in orientation/position of the sensor will lead to poor accuracy.

````python
def draw_lines_extra(img, lines, vertices, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    xp = []
    yp = []
    xn = []
    yn = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
#             if slope > 0:#0.55 and slope < 0.75:
            if x1 > vertices[0,0,2,0] and x2 < vertices[0,0,3,0]:
                xp += [x1,x2]
                yp += [y1,y2]
#             elif slope < 0:#-0.7 and slope > -0.8:
            elif x1 > vertices[0,0,0,0] and x2 < vertices[0,0,1,0]:
                xn += [x1,x2]
                yn += [y1,y2]
            else:
                continue
#             cv2.line(img, (x1, y1), (x2, y2), color, thickness)

    if bool(xp):
        poly_coeff_pos = np.polyfit(xp,yp,1)
#         print(poly_coeff_pos)
#         xp.sort()
        xp = [min(xp),max(xp)]
        y_pos = [int(np.polyval(poly_coeff_pos,i)) for i in xp]
        for i in range(len(xp)-1):
            cv2.line(img,(xp[i], y_pos[i]),(xp[i+1], y_pos[i+1]), color, thickness)           
    #         cv2.line(img,(xp[i], yp[i]),(xp[i+1], yp[i+1]), color, thickness)                 
    
    if bool(xn):
        poly_coeff_neg = np.polyfit(xn,yn,1)
#         xn.sort()
        xn = [min(xn),max(xn)]
        y_neg = [int(np.polyval(poly_coeff_neg,i)) for i in xn]        
        for i in range(len(xn)-1):
            cv2.line(img,(xn[i],y_neg[i]),(xn[i+1],y_neg[i+1]),color,thickness)
#             cv2.line(img,(xn[i],yn[i]),(xn[i+1],yn[i+1]),color,thickness)        
    
def hough_lines_extra(img, rho, theta, threshold, min_line_len, max_line_gap, vertices):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines_extra(line_img, lines, vertices)
    return line_img

def pipeline_extra(img):
    # 720x1280
    gray_scale = grayscale(img)
    kernel_size = 3
    blur_gray = cv2.GaussianBlur(gray_scale,(kernel_size, kernel_size),0)
    img_canny = canny(blur_gray,50,250)
#     vertices = np.array([[((150,550),(450,320),(650,320),(900,550))]], dtype=np.int32)
    vertices = np.array([[((200,720),(600,350),(700,350),(1200,720))]], dtype=np.int32)
#     vertices = np.array([[((200,720),(500,400),(800,450),(1200,720))]], dtype=np.int32)
#     vertices = np.array([[((220,720),(500,420),(800,460),(1200,720))]], dtype=np.int32)
    masked_img = region_of_interest(img_canny, vertices)
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = 0.125*np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 5#7    # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 #minimum number of pixels making up a line
    max_line_gap = 4    # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on
    img_hough = hough_lines_extra(masked_img, rho, theta, threshold, min_line_length, max_line_gap, vertices)
    final_img = weighted_img(img_hough, img, α=0.5, β=2., λ=0.5)
#     final_img = weighted_img(img_canny, img, α=0.5, β=2., λ=0.5)    
#     final_img = img_hough
    return final_img
````
